OUTLIER

```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
crop.columns
```

```python
wisker(crop["P"])
```

```python
for i in ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']:
    lw,uw=wisker(crop[i])
    crop[i]=np.where(crop[i]<lw,lw,crop[i])
    crop[i]=np.where(crop[i]>uw,uw,crop[i])
```

```python
for i in ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']:
    sns.boxplot(data=crop,x=i)
    plt.show()
```
